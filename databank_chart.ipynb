```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_bank = '/content/Data Bank.xlsx'
df = pd.read_excel(data_bank, sheet_name='Data')

print(df)
```

```python
count_0 = df[df['Personal Loan'] == 0].shape[0]
count_1 = df[df['Personal Loan'] == 1].shape[0]

labels = ['Not Accept', 'Accept']
values = [count_0, count_1]
```

```python
plt.bar(labels, values, color=['red', 'green'])
plt.title('Personal Loan Acceptance')
plt.xlabel('Acceptance Status')
plt.ylabel('Count')

plt.show()
```
